```python
import sys
sys.path.insert(0,"../pyphasefield-develop_sp/pyphasefield-develop_sp/pyphasefield/")
import pyphasefield.Engines as engines
import matplotlib.pyplot as plt
import numpy as np
import time
import math

def run_sim(x, ts, size):
    sim = engines.PCR_GPU_Engine_2(dimensions=[size, size])
    sim.set_framework("GPU_PARALLEL") 
    sim.set_dx(x)
    sim.set_dt(1e-10)
    sim.set_time_step_counter(0)
    sim.set_save_path(None)
    sim.set_autosave_flag(False)
    sim.set_autosave_save_images_flag(False)
    sim.set_autosave_rate(100000)
    sim.set_boundary_conditions(["NEUMANN", "NEUMANN"])

        #initialize non-array parameters
    sim.set_framework("GPU_SERIAL") #"CPU_SERIAL", "GPU_SERIAL"
    sim.set_dx(x)
    sim.set_dt(1e-10)
    sim.set_time_step_counter(0)
    #sim.set_temperature_type("ISOTHERMAL") #None, "ISOTHERMAL", "LINEAR_GRADIENT", "XDMF_FILE" ???
    #sim.set_temperature_initial_T(2294) #???
    #sim.set_temperature_dTdx(None) # temp grad
    #sim.set_temperature_dTdy(None)
    #sim.set_temperature_dTdz(None)
    sim.set_temperature_units("K")
    sim.set_save_path(None)
    sim.set_autosave_flag(False)
    sim.set_autosave_save_images_flag(False)
    sim.set_autosave_rate(100000)
    sim.set_boundary_conditions(["NEUMANN", "NEUMANN"])
    
    data = {
     
        "VM":10.765e-6,
        "R":8.314,
        "kini": 1.316,
        "c_0": 0.5, #NI compo in liquid 
        "r":1,# relax const
        "G":10e7, #temp grad
        "Vs":0.01, #solidification vel
        "dT":0, #undercooling
        "initXpos":400,
        "timeEnd":5000000,
        "sigma": 0.373

    }
    sigma = sim.data["sigma"]
    xi = 2.8*dx
    
    data['epsilon'] = (4/pi) * math.sqrt(xi*sigma)
    data['w'] = 2*sigma / (xi*1.5)
    data['Mphi'] = sigma * math.sqrt(2*w) / (epsilon**3 * 7*pi/64)
    data['xi'] = 2.8*dx
    #加入到字典中
    
    sim.set_user_data(data)
    sim.initialize_engine()
    
#run simulation
    for i in range(1):
        sim.plot_simulation(fields=[0, 1, 2, 3, 4], interpolation="nearest", save_images=False) #
        t0 = time.time()
        sim.simulate(2000) #time steps
    
    return sim
    
    
```
